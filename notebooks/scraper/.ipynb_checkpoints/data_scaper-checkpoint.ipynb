#### Create product scraper function

```python
def scrape_product(url):
    
    response = requests.get(url)
    
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Product Name
    product_name = soup.title.text
    
    # Ingredient Section
    short_section = soup.find("div", class_="ingredlist-short-like-section")
    
    ingredients = short_section.find_all("a")
    
    ingredient_names = [
        i.text.strip()
        for i in ingredients
        if "how to read" not in i.text.lower()
    ]
    
    return {
        "product_name": product_name,
        "ingredients": ingredient_names
    }
```

#### better scraper function

```python
def scrape_product(url):
    
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Failed to retrieve {url}")
        return None
    
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Product Name
    product_name = soup.title.text.replace(" ingredients (Explained)", "").strip()
    
    # Ingredient Section
    short_section = soup.find("div", class_="ingredlist-short-like-section")
    
    if short_section is None:
        print(f"No ingredient section found for {url}")
        return None
    
    ingredients = short_section.find_all("a")
    
    ingredient_names = [
        i.text.strip().lower()
        for i in ingredients
        if "how to read" not in i.text.lower()
    ]
    
    ingredient_string = ", ".join(ingredient_names)
    
    return {
        "product_name": product_name,
        "ingredients": ingredient_string,
        "url": url
    }
```

#### What is better: headers, status code check, cleaner product name, lowercasing ingredients, ingredient string, returning url

#### Multiple brands 

```python
def get_product_links(brand_url):

    response = requests.get(brand_url)
    soup = BeautifulSoup(response.text, "html.parser")

    links = []

    for tag in soup.find_all("a"):
        href = tag.get("href")

        if href and href.startswith("/products/"):
            links.append(
                "https://incidecoder.com" + href
            )

    return list(set(links))
```

```python
all_links = []

brands = [
    "https://incidecoder.com/brands/the-ordinary",
    "https://incidecoder.com/brands/cosrx",
    "https://incidecoder.com/brands/cerave"
]

for brand in brands:

    urls = get_product_links(brand)

    print(brand, len(urls))

    all_links.extend(urls)

product_links = list(set(all_links))

print("Total URLs:", len(product_links))
```

```python
import time
import random

all_products = []

for idx, url in enumerate(product_links, start=1):

    try:

        data = scrape_product(url)

        if data:
            all_products.append(data)

            print(
                f"[{idx}/{len(product_links)}] "
                f"{data['product_name']}"
            )

        time.sleep(random.uniform(1, 2))

    except Exception as e:

        print(f"Error: {url}")
        print(e)
```

```python
df = pd.DataFrame(all_products)

print(df.shape)
```

```python
df = df.drop_duplicates(
    subset=["product_name"]
)

print(df.shape)
```
